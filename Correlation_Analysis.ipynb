```python
import numpy as np # for numerical operatons
import pandas as pd # data manipulation and analysis
import matplotlib.pyplot as plt
import seaborn as sns # visualization libraries
from google.colab import drive
drive.mount('/content/drive')
```

```python
#loading a csv file

df_car = pd.read_csv('/content/drive/MyDrive/Datasets/ data.csv')
```

# Correlation Analysis

```python
#selecting relevant columns
columns = ['Engine HP', 'MSRP', 'Popularity', 'city mpg', 'highway MPG']
# computing the correlation metrix
correlation_matrix = df_car[columns].corr()
#printing the correlation matrix
print('Correlation Matrix:\n', correlation_matrix)
# plotting a heatmap for better visualization
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
#adding title
plt.title('Correlation Matrix of Selected Variables')
plt.show()
```
