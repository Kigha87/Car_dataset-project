```python
import numpy as np # for numerical operatons
import pandas as pd # data manipulation and analysis
import matplotlib.pyplot as plt
import seaborn as sns # visualization libraries
from google.colab import drive
drive.mount('/content/drive')
```

```python
#loading a csv file

df_car = pd.read_csv('/content/drive/MyDrive/Datasets/ data.csv')
```

# Visualizations

```python
#1 Trend in pricing (MSRP) vs Market Size (Vehicle Size)
plt.figure(figsize=(8, 6))
sns.barplot(x='Vehicle Size', y='MSRP', data=df_car, estimator=lambda x: sum(x) / len(x))
plt.xlabel('Vehicle Size')
plt.ylabel('Avg MSRP ($)')
plt.title('Trends i Pricing (MSRP) by Vehicle Size')
plt.xticks(rotation=45)
plt.show()
```

```python
#2 price MSRP Distribution for different Drivetrains Driven_Wheels
plt.figure(figsize=(10, 6))
sns.boxplot(x='Driven_Wheels', y='MSRP', data=df_car)
plt.xlabel('Driven_Wheels')
plt.ylabel('MSRP ($)')
plt.title('Price Distribution by Driven Wheels')
plt.yscale('log') #log scale is to manage large MSRP range
plt.grid(axis='y', alpha=0.7)
plt.show()
```

```python
#3 Relationship between Horsepower Engine HP and price MSRP
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Engine HP', y='MSRP', data=df_car, alpha=0.5, color='blue')
plt.xlabel('Engine HP')
plt.ylabel('MSRP ($)')
plt.title('Relationship between Engine HP and MSRP')
plt.grid(True, alpha=0.7)
plt.show()
```

```python
#4 MPG City and Highway trends by transmission type
mpg_trends = df_car.groupby('Transmission Type')[['city mpg', 'highway MPG']].mean()
#plotting the line plot
plt.figure(figsize=(10, 6))
plt.plot(mpg_trends.index, mpg_trends['city mpg'], marker='o', label='City MPG', color='blue')
plt.plot(mpg_trends.index, mpg_trends['highway MPG'], marker='s', label='Highway MPG', color='green')
# adding labels and title
plt.xlabel('Transmission Type')
plt.ylabel('Average MPG')
plt.title('Trends in City MPG and Highway MPG by Transmission Type')
plt.legend()
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.7)
plt.show()
```
